```python
plt.figure(figsize=(10, 6))
plt.hist(df['TotalPay'], bins=20, color='skyblue')
plt.title('Salaries Employee')
plt.xlabel('Salary')
plt.ylabel('Employee')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.pie(df['JobTitle'].value_counts(), labels=df['JobTitle'].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

```python
average_salary=df.groupby('JobTitle')['TotalPay'].mean()
average_salary
```

```python
correlation = df['TotalPay'].corr(df['TotalPayBenefits'])
print("Correlation between TotalPay and TotalPayBenefits:", correlation)
import matplotlib.pyplot as plt
plt.scatter(df['TotalPayBenefits'], df['TotalPay'])
plt.xlabel('TotalPayBenefits')
plt.ylabel('TotalPay')
plt.title('Scatter Plot of TotalPay vs TotalPayBenefits')
plt.show()
```
